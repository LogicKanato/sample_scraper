```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_csv('result.csv')
df.loc[df['Year'] == 2020]

# df.drop(columns=['Status'], inplace=True)
```

```python
df.rename({"Constituency name": "ConstituencyName"}, axis=1, inplace=True)
df.rename({"Constituency size": "ConstituencySize"}, axis=1, inplace=True)
df.rename({"Number of electors": "NumberOfElectors"}, axis=1, inplace=True)
df.rename({"1-Employees' salaries": "EmployeesSalary"}, axis=1, inplace=True)
df.rename({"2-Service Contracts": "ServiceContracts"}, axis=1, inplace=True)
df.rename({"Designated traveller": "DesignatedTraveller"}, axis=1, inplace=True)
df.rename({"Member's accommodation expenses": "AccommodationExpenses"}, axis=1, inplace=True)
df.rename({"Member's per diem expenses": "DiemExpenses"}, axis=1, inplace=True)
df.rename({"Member's secondary residence expenses": "SecondaryResidenceExpenses"}, axis=1, inplace=True)
df.rename({"4-Hospitality": "Hospitality"}, axis=1, inplace=True)
df.rename({"5-Gifts": "Gifts"}, axis=1, inplace=True)
df.rename({"6-Advertising": "Advertising"}, axis=1, inplace=True)
df.rename({"7-Printing": "Printing"}, axis=1, inplace=True)
df.rename({"Ten percenters": "TenPercenters"}, axis=1, inplace=True)
df.rename({"Other printing related expenses": "PrintingExpenses"}, axis=1, inplace=True)
df.rename({"8-Offices": "Offices"}, axis=1, inplace=True)
df.rename({"OfficeLeases-Insurance-Utilities": "OfficeLeasesInsuranceUtilities"}, axis=1, inplace=True)
df.rename({"Furniture-Furnishing-EquipmentPurchases": "FurnitureFurnishingEquipmentPurchases"}, axis=1, inplace=True)
df.rename({"Equipment rentals": "EquipmentRentals"}, axis=1, inplace=True)
df.rename({"Informatics-TelecommunicationEquipmentPurchases": "InformaticsTelecommunicationEquipmentPurchases"}, axis=1, inplace=True)
df.rename({"Telecommunication services": "TelecommunicationServices"}, axis=1, inplace=True)
df.rename({"Repairs-Maintenance": "RepairsMaintenance"}, axis=1, inplace=True)
df.rename({"Postage-CourierServices": "PostageCourierServices"}, axis=1, inplace=True)
df.rename({"Materials-Supplies": "MaterialsSupplies"}, axis=1, inplace=True)
df.to_csv('result.csv', index=False)
df.loc[df['Year'] == 2020]
```

```python

df['Name'] = df['Name'].apply(lambda x: x.split('\t\t')[0])
```

```python
df['Quarter'] = df['Quarter'].apply(lambda x: np.NaN if x == 4 else x)
```

```python
# df['Member'] = df['Member'].astype(float)
# df['Designated traveller'] = df['Designated traveller'].astype(float)
# df['Dependants'] = df['Dependants'].astype(float)
# df['Employees'] = df['Employees'].astype(float)
# df['Member\'s accommodation expenses'] = df['Member\'s accommodation expenses'].astype(float)
# df['Member\'s per diem expenses'] = df['Member\'s per diem expenses'].astype(float)
# df['Member\'s secondary residence expenses']= df['Member\'s secondary residence expenses'].astype(float)
# 
df["Member's per diem expenses"] = df["Member's per diem expenses"].apply(lambda x: 0 if math.isnan(x) else x)

df['3-Travel'] = df['Member'] + df['Designated traveller'] + df['Dependants'] + df['Employees'] + df['Member\'s accommodation expenses'] + df['Member\'s per diem expenses'] + df['Member\'s secondary residence expenses']
```

```python
df.to_csv('result.csv', index=False)
df.loc[df['Year'] == 2012]
```
